```python
import numpy as np
import pandas as pd
```

```python
class KNNRegressior:

    def __init__(self , K=3):
        self.k = K

    def fit(self , X , y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def _euclidean_dist(self , X1 , X2):
        return np.sqrt(np.sum((X1 - X2) ** 2))
    
    def predict_one(self , x):
        
        distances = [self._euclidean_dist(x , x_train) for x_train in self.X_train]

        sort_indices = np.argsort(distances)[:self.k]

        y_lable = [self.y_train[i] for i in sort_indices]

        y_avg = np.mean(y_lable)
        return y_avg

    def predict(self , X):

        y_pred = [self.predict_one(x) for x in X]
        return y_pred
```

```python
X_train = [
    [12 , 15],
    [17 , 25],
    [17 , 23],
    [19 , 27],
    [21 , 29]
]

y_train = [13 , 15 , 17 , 19 , 20]

X_test = [
    [18 , 21],
    [20, 23]
]

model = KNNRegressior()
model.fit(X_train , y_train)
model.predict(X_test)
```
