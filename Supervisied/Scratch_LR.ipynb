```python
import numpy as np
```

```python
class LinearRegression:

    def __init__(self, learning_rate = 0.01 , n_iter = 1000):
        self.bias = None
        self.weight = None
        self.lr = learning_rate 
        self.n_iter = n_iter

    def fit(self , X , y):
        m,n = X.shape

        self.bias = 0
        self.weight = np.zeros(n)

        for i in range(self.n_iter):
            y_pred = self.bias + (X @ self.weight)
    
            d_b = 1/m * np.sum(y_pred - y)
            d_w = 1/m * (X.T @ (y_pred - y))
    
            self.bias -= self.lr * d_b
            self.weight -= self.lr * d_w

    def predict(self , X):

        y_pred = self.bias + (X @ self.weight)
        return y_pred
```

```python
X = np.array([[1] ,[2] ,[3] ,[4] ,[5]])
y = np.array([1,4,6,8,10])

model = LinearRegression()
model.fit(X , y)
model.predict(X)
```
